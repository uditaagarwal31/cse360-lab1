Part 1: Line Trajectory

````python
def sense(x):
    return x
````

````python
def simulate(Δt, x, u):
    x += Δt * u
    return x
````

````python
def control(t, y):
    a = 4
    b = 2
    ux = 2
    uy = 1

    return array([ux, uy])
````

````python
tf = 10.
Δt = 0.01    # Time step
time = linspace(0.,tf, int(tf / Δt) + 1)  # Time interval


# Initial conditions
x = array([1., 1.])
x_log = [copy(x)]

for t in time:
    y = sense(x)
    u = control(t, y)
    x = simulate(Δt, x, u)
    x_log.append(copy(x))

x_log = array(x_log)
````

````python
grid()
plot(x_log[:,0], x_log[:,1])
````
